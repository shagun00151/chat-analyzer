First task that we are performing is data preprocessing in which we are converting our input text file into pandas dataframe

```python
# Importing libraries
import re 
import pandas as pd 
```

```python
f = open('chat.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
# can use regex.com to generate the re
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
dates = re.findall(pattern, data)
```

```python
dates
```

```python
 df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
df.head()
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
# Clean up the 'message_date' column by removing whitespaces and characters
df['message_date'] = df['message_date'].str.strip(' -')

# Define the input date format
input_date_format = "%d/%m/%y, %H:%M"

# Convert the 'message_date' column to datetime
df['message_date'] = pd.to_datetime(df['message_date'], format=input_date_format)

print(df)
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
    users = []
    messages = []
    for message in df['user_message']:
        entry = re.split('([\w\W]+?):\s', message)
        if entry[1:]:  # user name
            users.append(entry[1])
            messages.append(" ".join(entry[2:]))
        else:
            users.append('group_notification')
            messages.append(entry[0])
            
    df['user'] = users
    df['message'] = messages
    df.drop(columns=['user_message'], inplace=True)        
```

```python
    df['only_date'] = df['date'].dt.date
    df['year'] = df['date'].dt.year
    df['month_num'] = df['date'].dt.month
    df['month'] = df['date'].dt.month_name()
    df['day'] = df['date'].dt.day
    df['day_name'] = df['date'].dt.day_name()
    df['hour'] = df['date'].dt.hour
    df['minute'] = df['date'].dt.minute
```

```python
temp = df[df['user'] != 'group_notification']
temp = temp[temp['message'] != '<Media omitted>\n']
```

```python
f = open('stopwords.txt','r')
stop_words = f.read()
```

```python
# these code are used to count the frequency of each letter in chats
words = []
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```

```python
import emoji
emojis  = []
for message in df['message']:
    emojis.extend([c for c in message if c in emoji.UNICODE_EMOJI['en']])
```

```python
pd.DataFrame(Counter(emojis).most_common(len(Counter(emojis))))
```
